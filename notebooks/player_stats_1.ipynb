```python
# I can't find positions, but I think  we did this somewhere else.
league = requests.get(f'https://api.sleeper.app/v1/league/{league_id}')
league = league.json()

positions = [position for position in league['roster_positions'] if position != 'BN']

```

```python
# oh how convenient
print(len(positions))
```

```python
# let's go ahead and assume they're aligned
```

```python
# elite variable naming!
# zip() let's you loop through it all at once
for a, b, c in zip(positions, matchups[0]['starters'], matchups[0]['starters_points']):
    print(a,b,c)
```

```python
# nested for-loop
# it's a loop inside a loop
for team in matchups:
    for a,b,c in zip(positions, team['starters'], team['starters_points']):
        new_row = {'week': week, 'team': team['roster_id'], 'player_id': b, 'player_points': c, 'fantasy_position': a, 'active_flag': 1}
        print(new_row)
```

```python
# let's build our dataframe! and export the csv
player_stats_list = []

for team in matchups:
    for a,b,c in zip(positions, team['starters'], team['starters_points']):

        new_row = {'week': week, 'team': team['roster_id'], 'player_id': b, 'player_points': c, 'fantasy_position': a, 'active_flag': 1}
        player_stats_list.append(new_row)

player_stats = pd.DataFrame(player_stats_list)


# just in case
player_stats = player_stats.drop_duplicates()

player_stats.to_csv(f'raw_data/active_stats_week_{week}.csv', index=False)
```

```python

week_start = 1
week_end = 4

# triple nested for-loop
for week in range(week_start,week_end+1):
    

    matchups = requests.get(f'https://api.sleeper.app/v1/league/{league_id}/matchups/{week}')
    matchups = matchups.json()

    player_stats_list = []

    for team in matchups:
        for a,b,c in zip(positions, team['starters'], team['starters_points']):

            new_row = {'week': week, 'team': team['roster_id'], 'player_id': b, 'player_points': c, 'fantasy_position': a, 'active_flag': 1}
            player_stats_list.append(new_row)

    player_stats = pd.DataFrame(player_stats_list)


    # just in case
    player_stats = player_stats.drop_duplicates()

    player_stats.to_csv(f'raw_data/active_stats_week_{week}.csv', index=False)
```
